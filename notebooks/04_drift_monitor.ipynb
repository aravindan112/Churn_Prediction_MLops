```python
import pandas as pd
import numpy as np
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset, DataQualityPreset
from evidently import ColumnMapping

# Load original cleaned data
df = pd.read_csv('../data/processed/churn_cleaned.csv')

# Split into reference (train) and current (test) data
# Reference = what the model was trained on
# Current = new data coming in production
reference_data = df.iloc[:5634].drop(columns=['Churn'])
current_data = df.iloc[5634:].drop(columns=['Churn'])

print("Reference data shape:", reference_data.shape)
print("Current data shape:", current_data.shape)
```

```python
# Generate drift report
report = Report(metrics=[
    DataDriftPreset(),
    DataQualityPreset()
])

report.run(reference_data=reference_data, current_data=current_data)

# Save as HTML
import os
os.makedirs('../reports', exist_ok=True)
report.save_html('../reports/drift_report.html')

print("Drift report saved to reports/drift_report.html")
```
